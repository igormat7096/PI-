# **02. Objetivo do Projeto** # 


### Definir as Necessidades de Gerenciamento e Armazenamento de Informações. E a equipe de desenvolvimento viu que é fundamental entender os processos da loja de vestuário para determinar quais dados precisam ser armazenados. ###  

 

## **Algumas necessidades comuns incluem:** ##  
 ### 1. Gerenciamento de Estoque: Controlar as peças de roupa, quantidades, cores, tamanhos, etc. 
 ### 2. Gestão de Vendas: Registrar transações de vendas, itens vendidos, valores, data, etc. 
 ### 3. Controle de Clientes: Armazenar informações dos clientes, como nome, contato e histórico de compras. 
 ### 4. Fornecedores: Gerenciar informações dos fornecedores de roupas e acessórios. 
 ### 5. Funcionários: Manter registro dos vendedores, gerentes, e outros funcionários. 

 
---
### **_Dessa forma agregando agilidade ao negócio do cliente e possibilitando manter um controle da empresa e gerando relatórios de dados que podem ser usados para definir estratégias para o crescimento empresarial_.** ###

# **3. Requisitos Funcionais** #  

 ### _Os requisitos funcionais descrevem o que o sistema deve fazer._ ###

> ## RF1: O sistema deve permitir que os usuários façam login usando um nome de usuário e senha. 

> ## RF2: O sistema deve permitir que os usuários registrem novos produtos. 

> ## RF3: O sistema deve gerar relatórios mensais de vendas. 

> ## RF4: O sistema deve permitir a busca de produtos por nome ou categoria. 

> ## RF5: O sistema deve enviar notificações por e-mail para os usuários quando houver atualizações importantes.  


# **4.Requisitos Não Funcionais** #  

 ### __Os requisitos não funcionais descrevem como o sistema deve se comportar.__ ###

> ## RNF1: O sistema deve ter uma disponibilidade de 99,9% durante o horário comercial. 

> ## RNF2: O tempo de resposta para consultas de banco de dados deve ser inferior a 2 segundos.  

> ## RNF3: O sistema deve ser compatível com os principais navegadores (Chrome, Firefox, Safari). 

> ## RNF4: O sistema deve garantir a segurança dos dados do usuário, utilizando criptografia.

> ## RNF5: O sistema deve ser escalável para suportar até 10.000 usuários simultâneos.

# **Agora vamos realizar uma analíse dos dados:** # 
---

### Nessa fase vamos observar os maiores cancelamentos e os principais motivos de cancelamento. Além de ter uma visão do grupo de clientes que mais realiza compras! ###  

 

```python
import pandas as pd 

tabela = pd.read_csv("csv")

display(tabela)
```

```python
display(tabela.info())
```

```python
tabela = tabela.dropna()
display(tabela.info())
```

```python
display(tabela["cancelamento"].value_counts())
```

```python
display(tabela["cancelamento"].value_counts(normalize=True).map("{:.2%}".format))
```

```python
import plotly.express as px 

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelamento", width=600)
    grafico.show()
```

```python
tabela = tabela[tabela['cancelamento'] < 5]
tabela = tabela[tabela['cancelamento'] <= 20]
display(tabela)
display(tabela["cancelamento"].value_counts())
display(tabela['motivo_cancelamento'].value_counts(normalize=True).map("{:.2%}".format))
```
