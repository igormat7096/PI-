# **3. Requisitos Funcionais** #  

 ### _Os requisitos funcionais descrevem o que o sistema deve fazer._ ###

> ## RF1: O sistema deve permitir que os usurios faam login usando um nome de usurio e senha. 

> ## RF2: O sistema deve permitir que os usurios registrem novos produtos. 

> ## RF3: O sistema deve gerar relatrios mensais de vendas. 

> ## RF4: O sistema deve permitir a busca de produtos por nome ou categoria. 

> ## RF5: O sistema deve enviar notificaes por e-mail para os usurios quando houver atualizaes importantes.  


# **4.Requisitos No Funcionais** #  

 ### __Os requisitos no funcionais descrevem como o sistema deve se comportar.__ ###

> ## RNF1: O sistema deve ter uma disponibilidade de 99,9% durante o horrio comercial. 

> ## RNF2: O tempo de resposta para consultas de banco de dados deve ser inferior a 2 segundos.  

> ## RNF3: O sistema deve ser compatvel com os principais navegadores (Chrome, Firefox, Safari). 

> ## RNF4: O sistema deve garantir a segurana dos dados do usurio, utilizando criptografia.

> ## RNF5: O sistema deve ser escalvel para suportar at 10.000 usurios simultneos.

# **Agora vamos realizar uma analse dos dados:** # 
---

### Nessa fase vamos observar os maiores cancelamentos e os principais motivos de cancelamento. Alm de ter uma viso do grupo de clientes que mais realiza compras! ###  

 

```python
import pandas as pd 

tabela = pd.read_csv("csv")

display(tabela)
```

```python
display(tabela.info())
```

```python
tabela = tabela.dropna()
display(tabela.info())
```

```python
display(tabela["cancelamento"].value_counts())
```

```python
display(tabela["cancelamento"].value_counts(normalize=True).map("{:.2%}".format))
```

```python
import plotly.express as px 

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelamento", width=600)
    grafico.show()
```

```python
tabela = tabela[tabela['cancelamento'] < 5]
tabela = tabela[tabela['cancelamento'] <= 20]
display(tabela)
display(tabela["cancelamento"].value_counts())
display(tabela['motivo_cancelamento'].value_counts(normalize=True).map("{:.2%}".format))
```
